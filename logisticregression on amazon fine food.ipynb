```python
import pandas as pd
import numpy as np
import sqlite3
import seaborn as sns
import matplotlib.pyplot as plt
con=sqlite3.connect('database.sqlite')
filtered_data=pd.read_sql_query(
"""
    SELECT*
    FROM Reviews
    WHERE Score!=3
    """,con)

def partition(x):
    if x<3:
        return 0
    return 1

actualscore=filtered_data['Score']
posneg=actualscore.map(partition)
filtered_data['Score']=posneg

print(filtered_data)
```

```python

final=filtered_data.drop_duplicates(subset={"UserId","ProfileName","Time","Text"},keep="first",inplace=False)
final=final[final.HelpfulnessNumerator<=final.HelpfulnessDenominator]

df=pd.DataFrame(final)
final=df.sample(2000)
print(final)
```

```python
final['Score'].value_counts()
x1=final[final['Score']==1]
x2=final[final['Score']==0]
print(x1)
print("**************************************************************************")
print(x2)
```

```python
from sklearn.utils import resample
x1=resample(x1,replace=True,n_samples=2000,random_state=123)
x2=resample(x2,replace=True,n_samples=2000,random_state=123)
print(x1)
print("*************************************************************************************************")
print(x2)
```

```python
final=pd.concat([x1,x2])
print(final)
```

```python
final['Score'].value_counts()
```

```python
final=final.sort_values('Time',axis=0,ascending=True)
print(final)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf_idf_vect=TfidfVectorizer(analyzer='word',ngram_range=(1,2),stop_words='english')
final_tf_idf=tf_idf_vect.fit_transform(final['Text'].values)
final_tf_idf.get_shape()
```

```python
features=tf_idf_vect.get_feature_names()
print(features)
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler(with_mean=False)
final_tf_idf=ss.fit_transform(final_tf_idf)
final_matrix=final_tf_idf.todense()
print(final_tf_idf)
print("**************************************************************************************************")
print("**************************************************************************************************")
print((final_matrix))
```

```python
X_train,X_test=np.split(final_matrix,[int(0.7*(len(final_tf_idf)))]);
y_train,y_test=np.split(final['Score'],[int(0.7*len(final['Score']))]);
tuned_parameters=[{'C':[10**-3,10**-2,10**-1,10**0,10**1,10**2,10**3]}]
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
model=GridSearchCV(LogisticRegression(),tuned_parameters,scoring='f1',cv=10)
model.fit(X_train,y_train)
print(model.best_estimator_)
```

```python
print(model.score(X_test,y_test))
```
